```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# Create a 2-qubit circuit with 2 classical bits for measurement
qc = QuantumCircuit(2, 2)  # 2 qubits, 2 classical bits
qc.h(0)  # Hadamard gate on qubit 0
qc.cx(0, 1)  # CNOT gate (qubit 0 controls, qubit 1 target)
qc.measure([0, 1], [0, 1])  # Measure qubits 0 and 1 into classical bits 0 and 1

# Simulate
simulator = AerSimulator()
result = simulator.run(qc, shots=1000).result()
counts = result.get_counts(qc)
print(counts)
plot_histogram(counts)
```

This is an example of an entangled Bell State.
