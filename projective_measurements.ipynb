```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# Create a circuit with 2 qubits (1 system, 1 ancilla) and 1 classical bit
qc = QuantumCircuit(2, 1)

# Step 1: Prepare a test state on qubit 0 (e.g., |+⟩ = (|0⟩ + |1⟩)/√2)
qc.h(0)

# Step 2: Apply unitary to rotate X-basis to Z-basis (Hadamard on system qubit)
qc.h(0)

# Step 3: Entangle system (qubit 0) with ancilla (qubit 1) using CNOT
qc.cx(0, 1)

# Step 4: Apply inverse unitary (Hadamard again) to system qubit
qc.h(0)

# Step 5: Measure ancilla (qubit 1) in Z-basis
qc.measure(1, 0)

# Draw the circuit
qc.draw('mpl')


```

```python
# Simulate to verify
simulator = AerSimulator()
result = simulator.run(qc, shots=1000).result()
counts = result.get_counts(qc)
print(counts)
plot_histogram(counts)
```
