```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_state_city, plot_histogram
import matplotlib.pyplot as plt

# Create a 2-qubit circuit to generate an entangled state (Bell state)
qc = QuantumCircuit(2, 2)
qc.h(0)  # Hadamard on qubit 0
qc.cx(0, 1)  # CNOT to entangle qubits 0 and 1
# State: (|00⟩ + |11⟩)/√2, a 1st-order tensor with 4 components
```

```python
#Get the statevector (tensor)
state = Statevector(qc)
print("Statevector (tensor components):", state.data)
# Visualize the statevector as a cityscape plot
plot_state_city(state, title="2-Qubit Statevector (Tensor)")
```

```python
# Simulate measurements to get probabilities
qc.measure([0, 1], [0, 1])  # Measure both qubits
simulator = AerSimulator()
result = simulator.run(qc, shots=1000).result()
counts = result.get_counts(qc)

# Visualize measurement probabilities
plot_histogram(counts, title="Measurement Probabilities")
```
